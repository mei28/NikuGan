```
import os 
from PIL import Image
import numpy as np
from tqdm import tqdm

# オリジナルの画像がある場所
ORIGIN_DIR = '/content/drive/MyDrive/niku-gan/origin_img/'

# リサイズ後のフォルダ 
RESIZE_DIR = '/content/drive/MyDrive/niku-gan/resize_img/'

# リサイズのサイズ
resize = 256

if not os.path.isdir(RESIZE_DIR):
    print('make directory...')
    os.makedirs(RESIZE_DIR)


```

```
files = os.listdir(ORIGIN_DIR)

for file_name in tqdm(files):
    img = Image.open(os.path.join(ORIGIN_DIR,file_name)).convert("RGBA"); img.close

    tmp = np.array(img)
    
    mask = tmp[:,:,3]<240
    tmp[mask, 0] = 255 #R
    tmp[mask, 1] = 255 #G
    tmp[mask, 2] = 255 #B
    
    img = Image.fromarray(tmp[:,:,0:3])

    width, height = img.size

    if width == height:
        tmp = img
    elif width > height:
        tmp = Image.new('RGB',(width,width),(255,255,255))
        tmp.paste(img, (0, (width - height)//2))
    else:
        tmp = Image.new('RGB',(height,height),(255,255,255))
        tmp.paste(img,((height-width)//2,0))
    
    img_resize = tmp.resize((resize,resize),Image.BICUBIC)
    img_resize.save(os.path.join(RESIZE_DIR,file_name))
    print(f'resize: {file_name}')
```
